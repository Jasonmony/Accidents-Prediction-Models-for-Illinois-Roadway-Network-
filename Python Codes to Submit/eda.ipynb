```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("all_data_with_spd_limt.csv", index_col=0)
```

```python
df.head()
```

```python
from matplotlib.ticker import PercentFormatter
data = df['Surface_Category']
plt.hist(data, weights=np.ones(len(data)) / len(data),bins=20, alpha=0.5)

plt.title('Histogram for Surface Category')
plt.xlabel('Surface Category')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()
```

```python
df['rururb'].replace(to_replace =2,
                 value ="Urban", inplace= True)
```

```python
df['rururb'].replace(to_replace =1,
                 value ="Rural", inplace=True)
```

```python
data = df['rururb']
plt.hist(data, weights=np.ones(len(data)) / len(data),bins=3 ,alpha=1)

plt.title('Histogram for Rural vs Urban')
plt.xlabel('Rurul vs Urban')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()
```

```python
df.no_lanes.unique()
```

```python
data = df['no_lanes']
plt.hist(data, weights=np.ones(len(data)) / len(data),bins=30 ,alpha=1)

plt.title('Histogram for number of lanes')
plt.xlabel('Number of lanes')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()
```

```python
data = df['oneway']
plt.hist(data, weights=np.ones(len(data)) / len(data),bins=10 ,alpha=1)

plt.title('Histogram for oneway vs twoway')
plt.xlabel('oneway vs twoway')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()	
```
