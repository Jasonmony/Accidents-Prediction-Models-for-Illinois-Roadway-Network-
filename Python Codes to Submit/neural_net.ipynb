```python
X_test=test_df.iloc[:,1:-1]
y_test=test_df.iloc[:,-1]
```

```python
dummy_y_test=np_utils.to_categorical(y_test)
```

```python

# define baseline model
def baseline_model():
	model = Sequential()
	model.add(Dense(8, input_dim=12, activation='relu'))
	model.add(Dense(4, activation='softmax'))
	model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
	return model
```

```python
def baseline_model1():
  model=Sequential()
  model.add(Dense(12, input_dim=12, activation='relu'))
  model.add(Dense(8, activation='relu'))
  model.add(Dense(4, activation='softmax'))
  model.compile(loss= 'categorical_crossentropy', optimizer= 'adam', metrics =['accuracy'])
  return model
```

```python
model = Sequential()
model.add(Dense(50, input_dim=12, activation='relu'))
model.add(Dense(4, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, dummy_y, epochs=20, batch_size=50, verbose=0)
# evaluate the model
scores = model.evaluate(X_test, dummy_y_test, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
 
# serialize model to JSON
#model_json = model.to_json()
#with open("model.json", "w") as json_file:
#    json_file.write(model_json)
# serialize weights to HDF5
#model.save_weights("model.h5")
#print("Saved model to disk")'''
```

```python
estimator = KerasClassifier(build_fn=baseline_model1, epochs=20, batch_size=50, verbose=0)
kfold = KFold(n_splits=5, shuffle=True)
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
