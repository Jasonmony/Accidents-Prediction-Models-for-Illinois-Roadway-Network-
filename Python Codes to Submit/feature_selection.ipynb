```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import f_classif
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import confusion_matrix, roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
```

```python
#read data

df_06= pd.read_csv("Filtered_Merged_06.csv")
df_07= pd.read_csv("Filtered_Merged_07.csv")
df_08= pd.read_csv("Filtered_Merged_08.csv")
df_09= pd.read_csv("Filtered_Merged_09.csv")
df_10= pd.read_csv("Filtered_Merged_10.csv")
```

```python
df_06=df_06.drop(['trf_cntl'], axis=1)
df_07=df_07.drop(['trf_cntl'], axis=1)
df_08=df_08.drop(['trf_cntl'], axis=1)
df_09=df_09.drop(['trf_cntl'], axis=1)
df_10=df_10.drop(['trf_cntl'], axis=1)
```

```python
df_06.loc[(df_06.AccCount > 0) & (df_06.AccCount<= 30),'Accident']=1
df_06.loc[(df_06.AccCount > 30) & (df_06.AccCount<= 50),'Accident']=2
df_06.loc[(df_06.AccCount > 50),'Accident']=3
df_06.loc[(df_06.AccCount == 0),'Accident']=0
```

```python
df_07.loc[(df_07.AccCount > 0) & (df_07.AccCount<= 30),'Accident']=1
df_07.loc[(df_07.AccCount > 30) & (df_07.AccCount<= 50),'Accident']=2
df_07.loc[(df_07.AccCount > 50),'Accident']=3
df_07.loc[(df_07.AccCount == 0),'Accident']=0
```

```python
df_08.loc[(df_08.AccCount > 0) & (df_08.AccCount<= 30),'Accident']=1
df_08.loc[(df_08.AccCount > 30) & (df_08.AccCount<= 50),'Accident']=2
df_08.loc[(df_08.AccCount > 50),'Accident']=3
df_08.loc[(df_08.AccCount == 0),'Accident']=0
```

```python
df_09.loc[(df_09.AccCount > 0) & (df_09.AccCount<= 30),'Accident']=1
df_09.loc[(df_09.AccCount > 30) & (df_09.AccCount<= 50),'Accident']=2
df_09.loc[(df_09.AccCount > 50),'Accident']=3
df_09.loc[(df_09.AccCount == 0),'Accident']=0
```

```python
df_10.loc[(df_10.AccCount > 0) & (df_10.AccCount<= 30),'Accident']=1
df_10.loc[(df_10.AccCount > 30) & (df_10.AccCount<= 50),'Accident']=2
df_10.loc[(df_10.AccCount > 50),'Accident']=3
df_10.loc[(df_10.AccCount == 0),'Accident']=0
```

```python
X=pd.concat([df_06, df_07, df_08, df_09, df_10],ignore_index=True)
```

```python
y=X['Accident']
```

```python
X=X.drop(['Accident'],axis =1)
```

```python

X=X.drop(['Severity_PDO'],axis =1)
X=X.drop(['Severity_Inj_TypeA'],axis =1)
X=X.drop(['Severity_Inj_TypeB'],axis =1)
X=X.drop(['Severity_Inj_TypeC'],axis =1)

```

```python
X=X.drop(['AccCount'],axis =1)
```

```python
X=X.drop(['Severity_Unknown'],axis =1)
```

```python
X=X.drop(['Severity_Fatal'],axis =1)
```

```python
#X=X.drop(['Weather_Clear'],axis =1)
X=X.drop(['Weather_Rain'],axis =1)
X=X.drop(['Weather_Snow'],axis =1)
X=X.drop(['Weather_Unknown'],axis =1)
X=X.drop(['Weather_Other'],axis =1)
X=X.drop(['Weather_Fog'],axis =1)
X=X.drop(['Weather_Hail'],axis =1)
X=X.drop(['Weather_CrossWind'],axis =1)

```

```python
bestfeatures = SelectKBest(score_func=f_classif, k=15)
fit = bestfeatures.fit(X,y.values.ravel())
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns.to_list())
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(15,'Score'))  #print 10 best features
```

```python
model = LogisticRegression(solver='liblinear')
rfe = RFE(model, 15)
fit = rfe.fit(X, y.values.ravel())
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s \n" % fit.ranking_)

print("Important Features:  \n")
for i in range(len(fit.support_)):
  if fit.support_[i]== True:
    print(X.columns.to_list()[i])
```
